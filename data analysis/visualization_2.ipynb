matplotlib allowing overlapping of plot data so that we can create combined plots but the size of all data values should be same 

```python
a = np.arange(1,10,.25)
b = a ** 2                              # make sqaures of all values in a array and store in b
c = np.random.randint(1,10,a.size)
```

```python
a.size == b.size == c.size
```

```python
# set the graph style
print(plt.style.available)
```

```python
plt.style.use('ggplot')
```

#### how to make a graph
1. get the data âœ…
2. set the graph size
3. pass data in graph
4. set title, xlabel, ylabel for graph
5. add some setting for tick, legend, grid, etc. (optional)
5. show graph

```python
plt.figure(figsize=(10,5))
plt.plot(a, 'b*:',label="normal data")
plt.plot(b, 'g^-',label="squared values")
plt.plot(c, 'r.',label='random values')
plt.title("Overlapping Line Plot example")
plt.xlabel("number of values")
plt.ylabel("data points")
plt.legend()
plt.show()
```

## subplots

```python
fig ,ax = plt.subplots(nrows=1,ncols=3, figsize=(15,5))
ax[0].plot(a,'b*:',label="normal data")
ax[0].set_title('normal data')
ax[1].plot(b,'ro-')
ax[2].plot(c,'gs')
ax[1].set_title('squared data')
plt.show()
```

```python
font = {
    'family':'serif',
    'color':'darkred',
    'weight':'normal',
    'size':16,
}
```

```python
x = np.linspace(0,5,100)
y = np.cos(2 * np.pi * x) * np.exp(-x)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y,'ko--')
plt.title("Damped exponential decay", fontdict=font)
plt.xlabel("time in sec",fontdict=font)
plt.ylabel("voltage",fontdict=font)
plt.text(2, .4, '$\cos(2 \pi t) \exp(-t)$',fontdict=font)
plt.show()
```
