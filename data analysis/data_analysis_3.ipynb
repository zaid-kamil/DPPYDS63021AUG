```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pd.set_option('max_columns',None)
```

```python
df = pd.read_csv('cleaned_canada_immigration_data.csv',index_col='Country')
```

```python
df.head(2)
```

```python
years = [str(i) for i in range(1980,2014)]
df['total'] = df.loc[:,years].sum(axis=1)
```

```python
df.sort_values(by='total', ascending=False, inplace=True)
```

```python
df.head(2)
```

```python
df.loc['India',years].plot(
    figsize=(10,6),
    title="Indian Immigration to Canada from 1980-2013",
    linewidth=3,
    color='red',
    linestyle=':',
    marker='o',
    markersize=5,
    alpha=.5,
    markeredgecolor='black',
    )
plt.xlabel('years',fontsize=14)
plt.ylabel('num of immigrants',fontsize=14)
plt.show()
```

```python
df.loc[['India','China'],years].plot(
    figsize=(10,6),
    title="Indian Immigration to Canada from 1980-2013",
    linewidth=3,
    color='red',
    linestyle=':',
    marker='o',
    markersize=5,
    alpha=.5,
    markeredgecolor='black',
    )
plt.xlabel('years',fontsize=14)
plt.ylabel('num of immigrants',fontsize=14)
plt.show()
```
